```python
import pandas as pd

df = pd.read_csv('summary_statistics_a.csv')
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of mean temperatures with seaborn
plt.figure(figsize=(10, 6))
sns.histplot(df['mean_temp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Mean Temperatures', fontsize=14)
plt.xlabel('Mean Temperature (°C)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Average mean temperature
average_mean_temp = df['mean_temp'].mean()
print(f"Average mean temperature: {average_mean_temp:.2f}°C")

# Average temperature standard deviation
average_std_temp = df['std_temp'].mean()
print(f"Average temperature standard deviation: {average_std_temp:.2f}°C")

# Buildings with at least 50% of their area above 18ºC
buildings_above_18 = (df['pct_above_18'] >= 50).sum()
print(f"Number of buildings with at least 50% of their area above 18ºC: {buildings_above_18}")

# Buildings with at least 50% of their area below 15ºC
buildings_below_15 = (df['pct_below_15'] >= 50).sum()
print(f"Number of buildings with at least 50% of their area below 15ºC: {buildings_below_15}")
```
